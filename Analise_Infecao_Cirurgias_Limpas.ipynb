
# Análise de Infecção em Cirurgias Limpas

Este notebook faz uma análise baseada nos dados da ANAHP (2023), onde o percentual de infecção em cirurgias limpas nos hospitais brasileiros foi de 0,4%. Utilizaremos a distribuição binomial para modelar o número de infecções esperadas em uma instituição que realiza 300 cirurgias por mês.



## Distribuição Binomial

A distribuição binomial é usada para modelar o número de sucessos (neste caso, infecções) em um número fixo de tentativas (cirurgias), onde cada tentativa tem a mesma probabilidade de sucesso. A fórmula para calcular a probabilidade binomial é:

P(X = k) = (n! / (k! (n-k)!)) * p^k * (1-p)^(n-k)


Onde:
- \( P(X = k) \) é a probabilidade de ocorrer exatamente \( k \) infecções.
- \( n \) é o número de cirurgias (neste caso, 300).
- \( p \) é a probabilidade de uma infecção ocorrer (0.004, ou 0,4%).
- \( k \) é o número de infecções que queremos calcular.




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parâmetros da distribuição binomial
n = 300  # Número de cirurgias por mês
p = 0.004  # Probabilidade de infecção (0.4%)

# Valores possíveis de k (número de infecções)
k_values = np.arange(0, 11)  # Vamos analisar de 0 a 10 infecções

# Calculando a probabilidade de cada k usando a função de massa de probabilidade da binomial
binom_probs = binom.pmf(k_values, n, p)

# Criando o gráfico
plt.figure(figsize=(10, 6))
plt.bar(k_values, binom_probs, color='blue', alpha=0.7)

# Adicionando rótulos e título
plt.title("Distribuição Binomial: Probabilidade de Infecção em 300 Cirurgias Limpas")
plt.xlabel("Número de Infecções (k)")
plt.ylabel("Probabilidade")
plt.show()
```


## Interpretação do Gráfico

O gráfico acima mostra a distribuição das probabilidades para o número de infecções que podem ocorrer em um mês, dado que a instituição realiza 300 cirurgias limpas e a taxa de infecção é de 0,4%. O valor mais provável é que ocorram **0 infecções** ou **1 infecção** em um mês, dado que a probabilidade de infecção por cirurgia é muito baixa.



## Conclusão

Com base nos dados da ANAHP (2023), a probabilidade de uma infecção em cirurgias limpas é muito baixa (0,4%). A distribuição binomial foi usada para modelar a probabilidade de diferentes números de infecções ocorrerem em 300 cirurgias realizadas por mês. O resultado mais provável é que a instituição tenha de 0 a 1 infecção por mês, com uma chance muito baixa de ocorrerem mais de 2 infecções.
