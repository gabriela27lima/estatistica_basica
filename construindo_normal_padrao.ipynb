
# Distribuição Normal


### O que é a Distribuição Normal?
A **distribuição normal** é uma distribuição simétrica em forma de sino, onde os valores estão mais concentrados em torno da média. Em uma distribuição normal, a média, a mediana e a moda são iguais.

### O que é a Distribuição Normal Padrão?
A **distribuição normal padrão** é uma versão especial da distribuição normal, com:
- **Média** = 0
- **Desvio padrão** = 1

Essa distribuição é usada como referência para padronizar outras distribuições.

### O que é Densidade?
No gráfico abaixo, a **densidade** representa a probabilidade relativa de valores caírem dentro de diferentes intervalos. A área sob a curva de densidade é igual a 1, representando 100% da probabilidade.

A densidade permite visualizar a distribuição de forma contínua, enquanto o histograma mostra contagens discretas.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

mean = 0
std_dev = 1
n_samples = 1000
samples = np.random.normal(mean, std_dev, n_samples)

plt.figure(figsize=(8,6))
sns.histplot(samples, bins=30, kde=True, color='teal', stat="density", alpha=0.6)
plt.title('Distribuição Normal com Seaborn')
plt.xlabel('Valor')
plt.ylabel('Densidade')
plt.grid(True)

plt.show()
```
