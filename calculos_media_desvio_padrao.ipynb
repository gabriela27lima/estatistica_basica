```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Suponha que temos os seguintes valores:
media = 35  # Média
desvio_padrao = 2.4  # Desvio Padrão
```

```python
# Primeiro quartil (Q1)
q1 = media - 0.675 * desvio_padrao
print(f'1º Quartil (Q1): {q1}')

# Mediana (Q2), que é igual à média em uma distribuição normal
q2 = media
print(f'Mediana (Q2): {q2}')

# Terceiro quartil (Q3)
q3 = media + 0.675 * desvio_padrao
print(f'3º Quartil (Q3): {q3}')

# Intervalo Interquartil (IQR)
iqr = q3 - q1
print(f'Intervalo Interquartil (IQR): {iqr}')
```

```python
# Gerando valores para o eixo x (vamos pegar valores em torno da média)
x = np.linspace(media - 4*desvio_padrao, media + 4*desvio_padrao, 1000)

# Calculando a distribuição normal para esses valores
y = stats.norm.pdf(x, media, desvio_padrao)

# Criando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribuição Normal')
plt.fill_between(x, y, alpha=0.3, color='blue')
plt.title('Distribuição Normal ')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.axvline(media, color='red', linestyle='--', label='Média (μ)')
plt.axvline(media - desvio_padrao, color='orange', linestyle='--', label='1 Desvio Padrão (μ-σ)')
plt.axvline(media + desvio_padrao, color='orange', linestyle='--', label='1 Desvio Padrão (μ+σ)')
plt.legend()
plt.show()
```
